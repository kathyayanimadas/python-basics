```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
cars[['HP', 'WT','VOL','SP']]=sc.fit_transform(cars[['HP', 'WT','VOL','SP']])
```

```python
np.round(cars.describe(),2)
```

```python
#SIMPLE LINEAR REGRESSION MODEL----ols(ordinary least square)
ml_v=smf.ols('MPG~VOL',data = cars).fit()
ml_v.rsquared
```

```python
ml_w=smf.ols('MPG~WT',data = cars).fit()
ml_w.rsquared
```

```python
ml_v.params
```

```python
ml_sp=smf.ols('MPG~SP',data = cars).fit()
ml_sp.rsquared
```

```python
ml_hp=smf.ols('MPG~HP',data = cars).fit()
ml_hp.rsquared
```

```python
#prepare model
import statsmodels.formula.api as smf
model = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model.rsquared
```

```python
MPG= 0.359*WT + (-0.888)*VOL + 0.666*SP + (-1.348)*HP + 0.898
```
